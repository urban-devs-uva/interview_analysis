```python
#!pip install docx
#!pip install python-docx
from docx import Document
import matplotlib.pyplot as plt
import pandas as pd
import warnings
from bertopic import BERTopic
import re
import nltk
from nltk.corpus import stopwords
# nltk.download('stopwords')

warnings.filterwarnings("ignore")

doc = Document('nbhh.docx')
# preprocessing
stop_words = set(stopwords.words('english'))
preprocessed_doc = [
    ' '.join([word for word in re.findall(r'\b\w+\b', paragraph.text.lower()) if word not in stop_words])
    for paragraph in doc.paragraphs]
#print(preprocessed_doc)

paragraphs = []
current_paragraph = ''

for paragraph in doc.paragraphs:
    if paragraph.style.name.startswith('Heading'):
        if current_paragraph:
            paragraphs.append(current_paragraph.strip())
            current_paragraph = ''
    current_paragraph += paragraph.text + '\n'

if current_paragraph:
    paragraphs.append(current_paragraph.strip())

# Now paragraphs list contains each section based on headings
# display(paragraphs)
def analyze_topics(texts):
    model = BERTopic(verbose=True, embedding_model='paraphrase-MiniLM-L3-v2', min_topic_size=3)
    topics, _ = model.fit_transform(texts)
    
    freq = model.get_topic_info()
    print("Number of topics: {}".format(len(freq)))
    display(freq.head(60))
    return model, topics

# Extract text from the document
text_content = [paragraph.text for paragraph in doc.paragraphs if paragraph.text.strip()]

model, topics = analyze_topics(preprocessed_doc)  # Analyzing the list of paragraphs
freq = model.get_topic_info()


```

```python
import requests

def analyze_sentiment(text):
    url = "http://text-processing.com/api/sentiment/"
    data = {'text': text, 'language': 'english'}

    response = requests.post(url, data=data)
    result = response.json()

    sentiment = result['label']
    confidence = result['probability'][sentiment]

    return sentiment, confidence

freq['sentiment'], freq['confidence'] = zip(*freq['Representative_Docs'].apply(analyze_sentiment))

print(freq)
```
